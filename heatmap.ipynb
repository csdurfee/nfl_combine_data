```python
import get_combine_data
import pandas as pd
import numpy as np
import altair as alt
```

```python
all_data = get_combine_data.get_data(drafted_only=True)
```

```python
all_data.head()
```

```python
draft_pos_corr = get_combine_data.most_corr_with_draft_pos(all_data)

draft_pos_corr
```

```python
df = pd.DataFrame(draft_pos_corr).transpose()

df = df.reindex(sorted(df.columns), axis=1)

df.head()
```

```python
alt.Chart(df).mark_rect()
```

```python
unstacked = df.unstack().reset_index(name='value')

x = unstacked['level_0']
y = unstacked['level_1']
z = unstacked['value']

source = pd.DataFrame({'x': x.ravel(),
                     'y': y.ravel(),
                     'z': z.ravel()})

point = alt.selection_single(encodings=['x','y'])

chart = alt.Chart(source)

chart = chart.mark_rect().encode(
    x='x:O',
    y='y:O',
    color=alt.condition(point, 'z:Q', alt.value('lightgray'))
)

chart=chart.add_selection(point)


chart = chart.properties(width=300,height=300)

chart
```

above is correlation between performance vs others at that position and draft order.

so a high score for 'pos_d_40yd' 

players aren't drafted and then assigned to their position based on their body type
